###4.Customer Segmentation:
-- Segment customers based on their buying behavior, such as frequency of purchases and total spending.

```python
import pandas as pd

# Excel data
data = pd.read_excel('Superstore1.xlsx')

# total spending per customer
customer_spending = data.groupby('Customer ID')['Sales'].sum()

# frequency of purchases per customer
customer_frequency = data.groupby('Customer ID')['Order ID'].nunique()

# DataFrame with customer metrics
customer_metrics = pd.DataFrame({'Total Spending': customer_spending, 'Frequency': customer_frequency})

# segmentation
# segment customers into 3 groups based on quantiles
customer_metrics['Segment'] = pd.qcut(customer_metrics['Total Spending'], q=3, labels=['Low', 'Medium', 'High'])

# most valuable customer segments
most_valuable_segments = customer_metrics[customer_metrics['Segment'] == 'High']

# most valuable customer segments
print("Most Valuable Customer Segments:")
print(most_valuable_segments)
```

##Determine the optimal discount levels that maximize revenue while maintaining profitability.


```python
import numpy as np
```

```python
#range of discount levels (from 0% to 50%)
discount_levels = np.arange(0, 0.51, 0.05)

#revenue and profitability for each discount level
results = []

for discount in discount_levels:
    # discount to sales
    data['Discounted Sales'] = data['Sales'] * (1 - discount)

    # total revenue
    total_revenue = data['Discounted Sales'].sum()

    # total cost (assuming a constant cost percentage)
    total_cost = (data['Discounted Sales'] * 0.6).sum()  # Adjust 0.6 as needed

    # profitability
    profitability = total_revenue - total_cost

    results.append({
        'Discount': discount,
        'Total Revenue': total_revenue,
        'Total Cost': total_cost,
        'Profitability': profitability
    })

#storing the results
results_df = pd.DataFrame(results)

#profitability threshold 10% of total revenue)
profitability_threshold = 0.10 * results_df['Total Revenue'].max()

#optimal discount levels that meet or exceed the profitability threshold
optimal_discounts = results_df[results_df['Profitability'] >= profitability_threshold]

#optimal discount levels
print("Optimal Discount Levels:")
print(optimal_discounts)
```

##9. Customer Retention:
-- Analyze customer retention rates over time to understand customer loyalty.

```python
import pandas as pd
import matplotlib.pyplot as plt

#data
data = pd.read_excel('Superstore1.xlsx', sheet_name='Orders')

#'Order Date'
data['Order Date'] = pd.to_datetime(data['Order Date'])

# data by 'Customer ID' and first order date and latest order date
customer_activity = data.groupby('Customer ID').agg({'Order Date': ['min', 'max']})
customer_activity.columns = ['First Order Date', 'Latest Order Date']

#the retention rate for each time period (e.g., monthly)
retention_rate = []
periods = pd.date_range(start=customer_activity['First Order Date'].min(), end=customer_activity['Latest Order Date'].max(), freq='M')

for period in periods:
    active_customers = (customer_activity['First Order Date'] <= period) & (customer_activity['Latest Order Date'] >= period)
    retention_rate.append(active_customers.sum() / len(active_customers) * 100)

# the retention rates over time
plt.figure(figsize=(10, 6))
plt.plot(periods, retention_rate, marker='o', linestyle='-')
plt.xlabel('Time Period')
plt.ylabel('Retention Rate (%)')
plt.title('Customer Retention Rate Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

##Determine reorder points and lead times for efficient inventory management

```python
# Constants (Replace with actual data)
average_demand = 20  # Average daily demand for a product
desired_service_level = 0.95  # Desired service level (e.g., 95%)
standard_deviation_demand = 5  # Standard deviation of daily demand
lead_time = 7  # Lead time in days

# Z-score for the desired service level (e.g., 95% corresponds to Z = 1.65)
z_score = 1.65

#safety stock
safety_stock = z_score * standard_deviation_demand

#reorder point
reorder_point = (average_demand * lead_time) + safety_stock

#lead time demand
lead_time_demand = average_demand * lead_time

# Print results
print(f"Average Demand: {average_demand}")
print(f"Desired Service Level: {desired_service_level * 100}%")
print(f"Standard Deviation of Demand: {standard_deviation_demand}")
print(f"Lead Time: {lead_time} days")
print(f"Z-score: {z_score}")
print(f"Safety Stock: {safety_stock}")
print(f"Reorder Point: {reorder_point}")
print(f"Lead Time Demand: {lead_time_demand}")
```

20. Market Basket Analysis:
-- Use association rules to identify frequently co-purchased products.

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd


# encoding to create a binary matrix
basket = df.groupby(['Customer Name', 'Product Name'])['Quantity'].sum().unstack().fillna(0)

# Converting quantities to binary values (1 if purchased, 0 if not)
basket[basket > 0] = 1

# Applying Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True)

# Generating association rules
association_rules_df = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)

# Filtering association rules based on confidence and lift thresholds
filtered_rules = association_rules_df[(association_rules_df['confidence'] > 0.5) & (association_rules_df['lift'] > 1.0)]

#frequently co-purchased products and associated rules
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(filtered_rules)
```

###Generate recommendations for bundling products together to increase sales.

```python

#association rules by lift in descending order (higher lift indicates stronger association)
sorted_rules = filtered_rules.sort_values(by='lift', ascending=False)

#dictionary to store bundling recommendations
bundling_recommendations = {}

#sorted rules and identify products to bundle
for index, row in sorted_rules.iterrows():
    antecedent_products = row['antecedents']
    consequent_products = row['consequents']

    # Converting frozensets to lists
    antecedent_products = list(antecedent_products)
    consequent_products = list(consequent_products)

    #recommendation
    bundling_recommendations[f"Bundling {', '.join(antecedent_products)} with {', '.join(consequent_products)}"] = {
        'Antecedent Products': antecedent_products,
        'Consequent Products': consequent_products,
        'Lift': row['lift'],
        'Confidence': row['confidence']
    }

# recommendations
for recommendation, details in bundling_recommendations.items():
    print(recommendation)
    print("Antecedent Products:", details['Antecedent Products'])
    print("Consequent Products:", details['Consequent Products'])
    print("Lift:", details['Lift'])
    print("Confidence:", details['Confidence'])
    print()
```
